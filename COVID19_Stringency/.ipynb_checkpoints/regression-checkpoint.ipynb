```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np

from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils.fixes import loguniform
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
import pandas as pd
df = pd.read_csv('covid_preprocessed.csv', parse_dates=True)

print(df.shape)
```

```python
df.tail()
```

```python
# print(df.iloc[:, df.columns != ['date_x', 'state_id', 'deathIncrease']])

X = df.iloc[:, df.columns != 'deathIncrease']
y = df.iloc[:, 5]

X = df.drop(['date', 'state', 'deathIncrease'], axis=1)

print(X.tail())
print(y.tail())
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

fs = SelectKBest(score_func=f_regression, k=4)
# apply feature selection
X_selected = fs.fit(X, y)

#print(X_selected.shape)

df_scores = pd.DataFrame(X_selected.scores_)
df_columns = pd.DataFrame(X.columns)

print(df_scores)
print(df_columns)
```

```python
# concatenate dataframes
feature_scores = pd.concat([df_columns, df_scores],axis=1)
feature_scores.columns = ['Feature_Name','Score']  # name output columns
print(feature_scores.nlargest(4,'Score'))  # print 20 best features
# export selected features to .csv
df_univ_feat = feature_scores.nlargest(4,'Score')
#df_univ_feat.to_csv('feature_selection_UNIVARIATE.csv', index=False)
```

```python
X = df.drop(['date', 'state', 'deathIncrease', 'death', 'deathConfirmed', 'inIcuCurrently', 'totalTestsViral'], axis=1)
y = df.iloc[:, 5]

print(X.head())
print(y.head())
```

```python
param_dist = {'alpha': ['0.1', '0.25', '0.40', '0.5', '0.75', '1.0', '10', '50', '100', '500', '1000'],
              'fit_intercept': ['True', 'False'],
              'normalize': ['True', 'False'],
              'tol': loguniform(1e-4, 1e0),}

ridge_model = Ridge()

rrandom = RandomizedSearchCV(estimator=ridge_model, param_distributions=param_dist, cv=10, n_iter=100, random_state=42)
rrandom.fit(X, y)

print(rrandom.best_score_)
print(rrandom.best_estimator_.alpha)
print(rrandom.best_params_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)
```

```python
rfinal = Ridge(alpha= 0.25, fit_intercept=False, normalize=False, tol=0.5727904470799619)
rfinal.fit(X_train, y_train)
```

```python
pred = rfinal.predict(X_train)
```

```python
print(np.sqrt(mean_squared_error(y_train, pred, squared=False)))
print(r2_score(y_train, pred))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_train, pred)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=4)
ax.set_xlabel('Truths')
ax.set_ylabel('Predicted Values')
ax.set_title('COVID-19 Deaths - Ridge Regression')
plt.show()
ax.figure.savefig("ridgeregression.jpg", transparent=True, bbox_inches='tight')
```

```python
from sklearn.linear_model import ElasticNet
```

```python
param_dist = {"max_iter": [1, 5, 10, 100, 500],
                      "alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10, 100],
                      "l1_ratio": np.arange(0.0, 1.0, 0.1)}

enet_model = ElasticNet()

rrandom = RandomizedSearchCV(estimator=enet_model, param_distributions=param_dist, cv=10, n_iter=100, random_state=42)
rrandom.fit(X, y)

print(rrandom.best_score_)
print(rrandom.best_estimator_.alpha)
print(rrandom.best_params_)
```

```python
enetfinal = ElasticNet(max_iter=10, l1_ratio=0.2, alpha=100)
enetfinal.fit(X_train, y_train)
```

```python
pred = enetfinal.predict(X_train)

print(np.sqrt(mean_squared_error(y_train, pred, squared=False)))
print(r2_score(y_train, pred))
```

```python
from sklearn.svm import SVR

svr_model = SVR()

svr_model.fit(X_train, y_train)

pred = svr_model.predict(X_train)

print(np.sqrt(mean_squared_error(y_train, pred, squared=False)))
print(r2_score(y_train, pred))
```

```python
from sklearn.ensemble import BaggingRegressor

bag = BaggingRegressor()

bag.fit(X_train, y_train)

pred = bag.predict(X_train)

print(np.sqrt(mean_squared_error(y_train, pred, squared=False)))
print(r2_score(y_train, pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from numpy import mean
from numpy import std

cv = RepeatedKFold(n_splits = 5, n_repeats = 3, random_state=1)

scores = cross_val_score(bag, X, y, cv=cv)

#scores

print('MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
param_dist = {"n_estimators": [10, 25, 50, 100, 500, 1000]}

breg = BaggingRegressor()

rrandom = RandomizedSearchCV(estimator=breg, param_distributions=param_dist, cv=10, n_iter=100, random_state=42)
rrandom.fit(X, y)

print(rrandom.best_score_)
print(rrandom.best_estimator_)
print(rrandom.best_params_)
```

```python
bag = BaggingRegressor(n_estimators=50)

cv = RepeatedKFold(n_splits = 5, n_repeats = 3, random_state=1)

scores = cross_val_score(bag, X, y, cv=cv)

#scores

print('MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
from sklearn.ensemble import RandomForestRegressor

param_dist = {"n_estimators": [10, 25, 50, 100, 500, 1000],
             "max_depth": [1,5,10,25,50,100]}

rfr = RandomForestRegressor()

rrandom = RandomizedSearchCV(estimator=rfr, param_distributions=param_dist, cv=10, n_iter=100, random_state=42)
rrandom.fit(X, y)

print(rrandom.best_score_)
print(rrandom.best_estimator_)
print(rrandom.best_params_)
```

```python
rfr = RandomForestRegressor(max_depth=10, n_estimators=10)
cv = RepeatedKFold(n_splits = 5, n_repeats = 3, random_state=1)

scores = cross_val_score(rfr, X, y, cv=cv)

#scores

print('MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
from sklearn.kernel_ridge import KernelRidge

param_dist = {"alpha": [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]}

krr = KernelRidge()

rrandom = RandomizedSearchCV(estimator=krr, param_distributions=param_dist, cv=10, n_iter=100, random_state=42)
rrandom.fit(X, y)

print(rrandom.best_score_)
print(rrandom.best_estimator_.alpha)
print(rrandom.best_params_)
```

```python
krr = KernelRidge(alpha=100)
cv = RepeatedKFold(n_splits = 5, n_repeats = 3, random_state=1)

scores = cross_val_score(krr, X, y, cv=cv)

#scores

print('MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```

```python
from sklearn.ensemble import GradientBoostingRegressor

param_dist = {"n_estimators": [10, 25, 50, 100, 500, 1000]}

gbr = GradientBoostingRegressor()

rrandom = RandomizedSearchCV(estimator=gbr, param_distributions=param_dist, cv=10, n_iter=100, random_state=42)
rrandom.fit(X, y)

print(rrandom.best_score_)
print(rrandom.best_estimator_.alpha)
print(rrandom.best_params_)
```

```python
gbr = GradientBoostingRegressor(alpha=0.9, n_estimators=100)
cv = RepeatedKFold(n_splits = 5, n_repeats = 3, random_state=1)

scores = cross_val_score(gbr, X, y, cv=cv)

#scores

print('MAE: %.3f (%.3f)' % (mean(scores), std(scores)))
```
